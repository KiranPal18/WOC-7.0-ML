```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#import math
```

```python
def z_normalize(x):
  m,n = x.shape
  mean = np.zeros(n)
  std = np.zeros(n)
  for i in range(n):
    mean[i] = np.mean(x[:,i])
    std[i] = np.std(x[:,i])
    x[:,i] = (x[:,i] - mean[i])/std[i]
  return x
```

```python
def k_NN_data(path):
  path = path.strip("'")
  df = pd.read_csv(path)

  x = df[df.columns[1:-1]].to_numpy()
  y = df[df.columns[-1]].to_numpy()

  u , v = x.shape
  new = np.zeros(v)
  for i in range(v):
    new[i] = float(input("Enter the value of "+df.columns[i+1]+": "))

  return (x,y,new)
```

```python
x_knn , y_knn , new = k_NN_data('/content/binary_classification_train.csv')
```

```python
x_knn , y_knn , new = k_NN_data('/content/multi_classification_train.csv')
```

```python
def k_NN(x,y,new):
  u , v = x.shape
  p = int(input("Enter the value of order(p): "))

  distance = np.zeros((u,2))
  for i in range(u):
    distance[i,1] = 0
    for j in range(v):
      distance[i,1] += (x[i,j] - new[j])**p
    distance[i,1] = np.power(distance[i,1],(1.0/p))
    distance[i,0] = i

  distance_asc = np.argsort(distance[:,1])
  distance = distance[distance_asc]

  k = int(input("Enter the value of k: "))
  K = np.zeros(k)
  for i in range(k):
    K[i]= y[int(distance[i,0])]

  unique , freq = np.unique(K,return_counts=True)
  freq_count = np.zeros((len(unique),2))
  for i in range(len(unique)):
    freq_count[i,0] = unique[i]
    freq_count[i,1] = freq[i]
  print(f"Nearast neighbours: {freq_count}")
  class_predicted = freq_count[np.argmax(freq_count[:,1]),0]
  print(f"predicted class: {class_predicted}")
```

```python
k_NN(x_knn , y_knn , new)
```
