```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
def closest_centroids(X, centroids):
    K = centroids.shape[0]
    idx = np.zeros(X.shape[0], dtype=int)
    for i in range(X.shape[0]):
        distance = []
        for j in range(centroids.shape[0]):
            norm_ij = np.linalg.norm(X[i] - centroids[j])
            distance.append(norm_ij)

        idx[i] = np.argmin(distance)

    return idx
```

```python
def compute_centroids(X, idx, K):
    m, n = X.shape
    centroids = np.zeros((K, n))
    for k in range(K):
        points =X[idx == k]
        centroids[k] = np.mean(points, axis = 0)

    return centroids
```

```python
def kMeans_init_centroids(X, K):
    randidx = np.random.permutation(X.shape[0])
    centroids = X[randidx[:K]]

    return centroids
```

```python
def K_Means(X, initial_centroids, max_iters):
    m, n = X.shape
    K = initial_centroids.shape[0]
    centroids = initial_centroids
    previous_centroids = centroids
    idx = np.zeros(m)

    for i in range(max_iters):
        idx = closest_centroids(X, centroids)
        centroids = compute_centroids(X, idx, K)
    return centroids, idx
```

```python
def K_Means_data(path):
  path = path.strip("'")
  df = pd.read_csv(path)

  x = df.iloc[:,1:].to_numpy()
  return x
```

```python
x = K_Means_data('unsupervised_data.csv')
K = int(input("Enter the number of clusters: "))
num_itr = int(input("Enter the number of iteration: "))
initial_centroids = kMeans_init_centroids(x, K)
centroids, idx = K_Means(x, initial_centroids, num_itr)
print(f"position of centroids: {centroids}")
```
